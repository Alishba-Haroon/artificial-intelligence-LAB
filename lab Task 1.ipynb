# Hungman Game

```python
import random

hangman_image = [
    ''' 
  +---+
  |   |
      |
      |
      |
      |
=========''',
    ''' 
  +---+
  |   |
  O   |
      |
      |
      |
=========''',
    ''' 
  +---+
  |   |
  O   |
  |   |
      |
      |
=========''',
    ''' 
  +---+
  |   |
  O   |
 /|   |
      |
      |
=========''',
    ''' 
  +---+
  |   |
  O   |
 /|\  |
      |
      |
=========''',
    ''' 
  +---+
  |   |
  O   |
 /|\  |
 /    |
      |
=========''',
    ''' 
  +---+
  |   |
  O   |
 /|\  |
 / \  |
      |
========='''
]

words = 'listen sound semester guessed picture korean anime'.split()

def hangman_game():
    word = random.choice(words)
    missed = ''
    correct = ''
    lives = 6 

    while lives > 0 and '_' in ['_' if letter not in correct else letter for letter in word]:
        print(hangman_image[len(missed)])
        print('Missed letters:', ' '.join(missed))
        print('Lives remaining:', lives)
        blanks = ['_' if letter not in correct else letter for letter in word]
        print(' '.join(blanks))
        guess = input('Guess a letter: ').lower()

        if len(guess) != 1 or guess not in 'abcdefghijklmnopqrstuvwxyz':
            print('Please enter a single letter.')
        elif guess in missed or guess in correct:
            print('You already guessed that letter.')
        elif guess not in word:
            missed += guess
            lives -= 1 
        else:
            correct += guess

        if '_' not in blanks:
            print('You won! The word was', word)
            break
        elif lives == 0:
            print(hangman_image[len(missed)])
            print('You lost! The word was', word)
            break

    playAgain = input('Play again? (yes/no): ')
    if playAgain.lower().startswith('y'):
        hangman_game()

hangman_game()
```
